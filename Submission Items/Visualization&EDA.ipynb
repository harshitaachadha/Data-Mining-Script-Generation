```python
data=pd.read_csv('dataset.csv')
data.head()
```

```python
# Remove columns that are unnamed
data = data.loc[:, ~data.columns.str.contains('^Unnamed')]
```

```python

# Set the aesthetics for the plots
sns.set(style="whitegrid")

# Plotting the distribution of the 'vote_average'
plt.figure(figsize=(10, 6))
sns.histplot(data['vote_average'], kde=True)
plt.title('Distribution of Vote Average')
plt.xlabel('Vote Average')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(data.corr(), annot=False, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='holiday_episode', y='vote_average', data=data)
plt.title('Vote Average for Holiday Episodes')
plt.xlabel('Holiday Episode')
plt.ylabel('Average Vote')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Total Spoken Words', y='vote_average', data=data)
plt.title('Vote Average vs Total Spoken Words')
plt.xlabel('Total Spoken Words')
plt.ylabel('Vote Average')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sarcasm Index', y='vote_average', data=data)
plt.title('Vote Average vs Sarcasm Index')
plt.xlabel('Total Spoken Words')
plt.ylabel('Vote Average')
plt.show()
```

```python
numerical_vars = ['writer_popularity', 'director_popularity']

# Creating scatter plots
plt.figure(figsize=(18, 6))

for i, var in enumerate(numerical_vars, 1):
    plt.subplot(1, len(numerical_vars), i)
    sns.scatterplot(x=var, y='vote_average', data=data)
    plt.title(f'Vote Average vs {var}')
    plt.xlabel(var)
    plt.ylabel('Vote Average')

plt.tight_layout()
plt.show()
```

```python
# Plotting a line graph for vote_average vs air_date
plt.figure(figsize=(14, 7))
plt.plot(data['sentiment'].sort_values(), data['vote_average'], marker='', linestyle='-', color='blue')
plt.title('Vote Average vs Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Vote Average')
plt.grid(True)
plt.show()
```

```python
# Plotting a line graph for vote_average vs air_date
plt.figure(figsize=(14, 7))
plt.plot(data['air_date'].sort_values(), data['vote_average'], marker='', linestyle='-', color='blue')
plt.title('Vote Average vs Air Date')
plt.xlabel('Air Date')
plt.ylabel('Vote Average')
plt.grid(True)
plt.show()
```

```python
# Scatter plots for numerical variables
# Selecting a couple of numerical variables
numerical_vars = ['Rachel_spoken_words', 'Ross_spoken_words', 'Monica_spoken_words', 'Phoebe_spoken_words','Joey_spoken_words']

# Creating scatter plots
plt.figure(figsize=(18, 6))

for i, var in enumerate(numerical_vars, 1):
    plt.subplot(1, len(numerical_vars), i)
    sns.scatterplot(x=var, y='vote_average', data=data)
    plt.title(f'Vote Average vs {var}')
    plt.xlabel(var)
    plt.ylabel('Vote Average')

plt.tight_layout()
plt.show()
```

```python
# Pairplot for selected numerical attributes
numerical_attrs_pairplot = ['Total Spoken Words', 'number of scene changes', 'vote_average', 'Sarcasm Index']
sns.pairplot(data[numerical_attrs_pairplot])
plt.suptitle('Pairplot of Selected Numerical Attributes', y=1.02)
plt.show()
```

```python
# Bar Plot for 'holiday_episode'
plt.figure(figsize=(8, 5))
sns.countplot(x='holiday_episode', data=data)
plt.title('Count of Holiday Episodes')
plt.xlabel('Holiday Episode')
plt.ylabel('Count')
plt.show()

# Box Plot for 'vote_average' Grouped by Emotional Archetypes
# Selecting a few emotional archetypes for demonstration
emotional_archetypes = ['Emotional_Archetype_Cinderella', 'Emotional_Archetype_Icarus', 'Emotional_Archetype_Man in a Hole']

plt.figure(figsize=(12, 6))
data_melted = data.melt(id_vars='vote_average', value_vars=emotional_archetypes)
sns.boxplot(x='variable', y='vote_average', hue='value', data=data_melted)
plt.title('Vote Average Grouped by Emotional Archetypes')
plt.xlabel('Emotional Archetype')
plt.ylabel('Vote Average')
plt.xticks(rotation=45)
plt.show()

# Scatter Plot: 'director_popularity' vs. 'vote_average'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='director_popularity', y='vote_average', data=data)
plt.title('Director Popularity vs Vote Average')
plt.xlabel('Director Popularity')
plt.ylabel('Vote Average')
plt.show()
```

```python
radar_data = pd.DataFrame()
for archetype in emotional_archetypes:
    avg_spoken_words = data[data[archetype] == 1][characters].mean()
    radar_data = radar_data.append(avg_spoken_words, ignore_index=True)

# Plotting the Radar Chart
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Calculate the angle for each category
angles = [n / float(num_vars) * 2 * pi for n in range(num_vars)]
angles += angles[:1]

# Plot each archetype
for i in range(len(radar_data)):
    values = radar_data.iloc[i].tolist()
    values += values[:1]
    ax.plot(angles, values, linewidth=2, linestyle='solid', label=emotional_archetypes[i])
    ax.fill(angles, values, alpha=0.4)

# Add labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(characters)

# Add a legend and title
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.title('Average Spoken Words per Character Across Emotional Archetypes')

plt.show()

# Violin Plot for 'Total Spoken Words' across different emotional archetypes
plt.figure(figsize=(12, 6))
data_melted = data.melt(id_vars=['Total Spoken Words'], value_vars=emotional_archetypes, var_name='Emotional Archetype', value_name='Archetype Present')
sns.violinplot(x='Emotional Archetype', y='Total Spoken Words', hue='Archetype Present', data=data_melted, split=True)
plt.title('Distribution of Total Spoken Words Across Emotional Archetypes')
plt.xlabel('Emotional Archetype')
plt.ylabel('Total Spoken Words')
plt.xticks(rotation=45)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# 3D Scatter Plot for 'director_popularity', 'writer_popularity', and 'vote_average'
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

x = data['director_popularity']
y = data['writer_popularity']
z = data['vote_average']

ax.scatter(x, y, z, c=z, cmap='viridis', marker='o')

ax.set_xlabel('Director Popularity')
ax.set_ylabel('Writer Popularity')
ax.set_zlabel('Vote Average')

plt.title('3D Scatter Plot: Director and Writer Popularity vs Vote Average')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Hexbin Plot for 'director_popularity' and 'writer_popularity'
plt.figure(figsize=(10, 6))
plt.hexbin(data['director_popularity'], data['writer_popularity'], gridsize=30, cmap='Blues', mincnt=1)
plt.colorbar(label='Count in Bin')
plt.title('Hexbin Plot of Director Popularity vs Writer Popularity')
plt.xlabel('Director Popularity')
plt.ylabel('Writer Popularity')
plt.show()
```
