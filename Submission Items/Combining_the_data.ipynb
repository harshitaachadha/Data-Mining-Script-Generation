# Combining Meta data and Engineered features

At first I import Irfan's part of data hat contains scraped metadata, character sets, overall sentiment, etc.

```python
import pandas as pd
df_meta = pd.read_csv('preprocessed3.csv')
df_meta.drop(columns = ['air_date','title_len','Overview_sentiment','Ross', 'Monica', 'Chandler', 'Joey', 'Phoebe',
       'Rachel', 'Gunther', 'Geller', 'Director', 'Mike', 'Janice', 'Carol','Susan'], inplace=True)
df_meta.head()
```

Next I import a csv file to get id in script data

```python
df_id = pd.read_csv('episodes_final_df.csv')
df_id.drop(columns = ['air_date','name','overview','runtime','vote_average','list_of_characters','average_character_popularity'], axis=1,inplace=True)
df_id.head()
```

Importing the scripts data

```python
df_script = pd.read_csv('Final_Script_Feature_Engineered.csv')
df_script.drop(columns = ['Unnamed: 0.1','Unnamed: 0'], axis=1,inplace=True)
df_script.head()
```

```python
df_script = df_script.dropna(subset=['Episode'])
```

```python
df_script['Episode'] = df_script['Episode'].astype(int)
df_script['Season'] = df_script['Season'].astype(int)
df_script.head()
```

```python
merged_df = pd.merge(df_id, df_script, left_on=['episode_number', 'season_number'], right_on=['Episode', 'Season'], how='inner')
merged_df.head()
```

```python
merged_df.drop(columns = ['episode_number',	'season_number'], axis=1,inplace=True)
merged_df.head()
```

```python
final_merged_df = pd.merge(merged_df, df_meta, on='id', how='inner')
final_merged_df.head()
```

```python
final_merged_df.columns
```
